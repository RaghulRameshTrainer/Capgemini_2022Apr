```python
import matplotlib.pyplot as plt
plt.plot([5,8,10],[11,16,6],'g',label="Line1",linewidth=3)
plt.plot([6,9,11],[6,15,7],'c',label="Line2",linewidth=3)
plt.title("Sample-Diagram")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

```python
plt.bar([1,3,5,7,9],[5,2,7,8,2],label="2021-Sales",width=0.7)
plt.bar([2,4,6,8,10],[8,6,2,5,6],color='g',label="2022-Sales",width=0.7)
plt.title("Sample-Diagram")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()

plt.show()
```

```python
hrs=[7,4,4,9]
activity=['sleeping','eating','working','playing']
cols=['c','m','r','b']

plt.pie(hrs,
       labels=activity,
       colors=cols,
       startangle=90,
       shadow=True,
       autopct='%1.1f%%',
       explode=(0,0,0.1,0))
plt.show()
```
